```python
# Import pandas
import pandas as pd
```

```python
# Load spreadsheet
xl = pd.ExcelFile("../Input/government_shutdown_stats.xlsx")

# Load a sheet into a DataFrame by name:
df = xl.parse(xl.sheet_names[0])

# # Print the df
# print(shutdown_df)
# print()
```

```python
print("Let's briefly explore our Data Frame:")
print("-------------------------------------")
print()
print(f"The shape (rows, columns) of our data frame is {df.shape}")
print()
print(f"The number of columns is {len(df.columns)}")
print()
print(f"The number of rows is {len(df)}")
print()
print(f"The number of rows is {format(df.shape[0], ',')} & the number of columns is {df.shape[1]}")
print()
print(f"The {len(df.columns)} columns are labeled as follows:")
print(f"-------------------------------------")
print(f"{df.columns}")
print()
print(f"Making them easier to read, that is:")
print(f"------------------------------------")
for col in df.columns:
        print(col)
print()
print(f"The data types of all items in out data frame are as follows:")
print(f"-------------------------------------------------------------")
print(f"{df.dtypes}")
print()
print("Let's describe our numeric data:")
print("--------------------------------")
print(df.describe())
print()
print("Let's check for null data:")
print("--------------------------")
print(df.count())
print()
print(df.isnull().any())
```
